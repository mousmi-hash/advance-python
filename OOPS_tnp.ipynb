```python

def vehicle_entry(parking, capacity):
    if len(parking) >= capacity:
        print("Parking Lot Full")
    else:
        vehicle = input("Enter Vehicle Number: ")
        hours = int(input("Enter expected parking hours: "))
        parking[vehicle] = hours
        print("Vehicle Entered Successfully")

def vehicle_exit(parking, rate):
    vehicle = input("Enter Vehicle Number: ")

    if vehicle in parking:
        hours = parking[vehicle]
        fee = hours * rate
        print("Parking Time:", hours, "hours")
        print("Parking Fee: ₹", fee)
        del parking[vehicle]
    else:
        print("Vehicle not found")

def available_spots(parking, capacity):
    spots = capacity - len(parking)
    print("Available Spots:", spots)

def main():
    capacity = 5
    rate = 20
    parking = {}

    while True:
        print("\n1. Vehicle Entry")
        print("2. Vehicle Exit")
        print("3. Available Spots")
        print("4. Exit")

        choice = int(input("Enter choice: "))

        if choice == 1:
            vehicle_entry(parking, capacity)
        elif choice == 2:
            vehicle_exit(parking, rate)
        elif choice == 3:
            available_spots(parking, capacity)
        elif choice == 4:
            print("Program Ended")
            break
        else:
            print("Invalid Choice")


main()
```

```python
class Task:
    def __init__(self, name, category, deadline, priority):
        self.name = name
        self.category = category
        self.deadline = deadline
        self.priority = priority

tasks = []

def add_task():
    name = input("Task Name: ")
    category = input("Category: ")
    deadline = input("Deadline: ")
    priority = input("Priority (High/Medium/Low): ")
    tasks.append(Task(name, category, deadline, priority))
    print("Task Added Successfully!")

def view_tasks():
    for t in tasks:
        print(f"{t.name} | {t.category} | {t.deadline} | {t.priority}")

def filter_category(cat):
    for t in tasks:
        if t.category == cat:
            print(f"{t.name} | {t.deadline} | {t.priority}")
add_task()
view_tasks()
filter_category("Study")
```

```python

def billing_system():
    cart = {}
    total = 0

    n = int(input("Enter number of products: "))

    for i in range(n):
        name = input("Enter product name: ")
        price = float(input("Enter price: "))
        qty = int(input("Enter quantity: "))
        cart[name] = price * qty
        total += price * qty

    discount = float(input("Enter discount %: "))
    final = total - (total * discount / 100)

    print("\nBill Summary")
    for item in cart:
        print(item, ":", cart[item])

    print("Total:", total)
    print("Final Amount:", final)

billing_system()
```

```python

def movie_booking():
    seats = 50
    movies = ["Avengers", "Pushpa", "KGF"]

    print("Available Movies:")
    for m in movies:
        print(m)

    movie = input("Choose movie: ")
    tickets = int(input("Enter number of tickets: "))

    if tickets <= seats:
        price = tickets * 150
        print("\nBooking Confirmed")
        print("Movie:", movie)
        print("Tickets:", tickets)
        print("Total Price:", price)
        seats -= tickets
    else:
        print("Not enough seats")

movie_booking()
```

```python

def calculate_grade_point(mark):
    if mark >= 90:
        return 10
    elif mark >= 80:
        return 9
    elif mark >= 70:
        return 8
    elif mark >= 60:
        return 7
    elif mark >= 50:
        return 6
    elif mark >= 40:
        return 5
    else:
        return 0



def calculate_cgpa(marks):
    total_points = 0
    for mark in marks:
        total_points += calculate_grade_point(mark)
    cgpa = total_points / len(marks)
    return cgpa


name = input("Enter Student Name: ")
n = int(input("Enter number of subjects: "))

marks = []

for i in range(n):
    m = int(input(f"Enter marks for subject {i+1}: "))
    marks.append(m)

cgpa = calculate_cgpa(marks)

print("\nStudent Name:", name)
print("Marks:", marks)
print("CGPA:", round(cgpa, 2))7
```

```python
import hashlib

passwords = {}

def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

def add_password():
    site = input("Enter website: ")
    pwd = input("Enter password: ")
    passwords[site] = hash_password(pwd)
    print("Password stored successfully!")

def view_passwords():
    for site, pwd in passwords.items():
        print(site, ":", pwd)

while True:
    print("\n1. Add Password")
    print("2. View Passwords")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_password()
    elif choice == "2":
        view_passwords()
    elif choice == "3":
        break
    else:
        print("Invalid choice")
```

```python
freelancers = []
clients = []
projects = []

def add_freelancer():
    name = input("Enter freelancer name: ")
    freelancers.append(name)

def add_client():
    name = input("Enter client name: ")
    clients.append(name)

def assign_project():
    client = input("Client name: ")
    freelancer = input("Freelancer name: ")
    project = input("Project title: ")

    projects.append((client, freelancer, project))
    print("Project assigned!")

def view_projects():
    for p in projects:
        print("Client:", p[0], "| Freelancer:", p[1], "| Project:", p[2])

while True:
    print("\n1.Add Freelancer\n2.Add Client\n3.Assign Project\n4.View Projects\n5.Exit")

    ch = input("Choice: ")

    if ch == "1":
        add_freelancer()
    elif ch == "2":
        add_client()
    elif ch == "3":
        assign_project()
    elif ch == "4":
        view_projects()
    elif ch == "5":
        break
```

```python
inventory = {}

def add_item():
    item = input("Enter item name: ")
    qty = int(input("Enter quantity: "))
    inventory[item] = inventory.get(item, 0) + qty

def remove_item():
    item = input("Enter item name: ")
    qty = int(input("Enter quantity to remove: "))

    if item in inventory and inventory[item] >= qty:
        inventory[item] -= qty
    else:
        print("Not enough stock")

def show_inventory():
    for item, qty in inventory.items():
        print(item, ":", qty)

while True:
    print("\n1.Add Item\n2.Remove Item\n3.Show Inventory\n4.Exit")

    ch = input("Enter choice: ")

    if ch == "1":
        add_item()
    elif ch == "2":
        remove_item()
    elif ch == "3":
        show_inventory()
    elif ch == "4":
        break
```
